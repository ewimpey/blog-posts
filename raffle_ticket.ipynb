```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Define the dimensions of the heatmap
num_your_contributions = 163
num_other_contributions = 300

# Create a grid of all possible combinations of your contributions and other contributions
your_contributions = np.arange(1, num_your_contributions + 1)
other_contributions = np.arange(1, num_other_contributions + 1)
your_contribution_grid, other_contribution_grid = np.meshgrid(your_contributions, other_contributions)


```

```python
def heatmap_function(your_contribution, other_contribution, prize_value):
    total_contribution = your_contribution + other_contribution
    prob_win = your_contribution/total_contribution
    ev = (prob_win * prize_value) - your_contribution
    return ev
```

```python
# Apply the heatmap function to each combination of contributions to get the corresponding heatmap value
prize = 163
heatmap_values = heatmap_function(your_contribution_grid, other_contribution_grid, prize)

# Create the heatmap using seaborn
sns.set()
heatmap = sns.heatmap(heatmap_values, cmap="YlGnBu")

# Invert the y-axis to reverse the order of the labels
heatmap.invert_yaxis()

# Show the plot
plt.show()
```

```python
# Apply the heatmap function to each combination of contributions to get the corresponding heatmap value
prize = 163
heatmap_values = heatmap_function(your_contribution_grid, other_contribution_grid, prize)

# Create the heatmap using seaborn
sns.set()
cmap = sns.diverging_palette(20, 220, s=80, center='light')
heatmap = sns.heatmap(heatmap_values, cmap=cmap, center=0)
# Invert the y-axis to reverse the order of the labels
heatmap.invert_yaxis()

heatmap.set(xlabel='Amount you contributed', ylabel='Amount others contributed')

# Format the color bar to show dollar amounts
colorbar = heatmap.collections[0].colorbar
colorbar.set_ticks(np.arange(-150, 150, 50))
colorbar.set_ticklabels(['${:.0f}'.format(x) for x in np.arange(-150, 150, 50)])
colorbar.ax.set_ylabel('Expected Value')

# Add a title to the plot
heatmap.set_title('Raffle Ticket Contribution Guide ($162 Prize Value)')

# Show the plot
plt.show()
```
